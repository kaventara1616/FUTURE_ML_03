```python
with open("job_description.txt","r") as file:
    job = file.read()

print(job)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
texts = [job] + resume["Resume"].tolist()

vectorizer = TfidfVectorizer()

matrix = vectorizer.fit_transform(texts)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
scores = cosine_similarity(matrix[0:1], matrix[1:]).flatten()

resume["Score"] = scores
```

```python
resume = resume.sort_values(by="Score", ascending=False)

resume
```

```python
job_skills = ["python","sql","machine learning","pandas"]
```

```python
def missing_skills(text):

    text = text.lower()

    missing = []

    for skill in job_skills:
        if skill not in text:
            missing.append(skill)

    return missing
```

```python
resume["Missing Skills"] = resume["Resume"].apply(missing_skills)
```
